```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("insurance.csv")
df=data.copy()
df.head()
```

```python
df.isna().sum()
```

```python
df.hist(bins=10, figsize=(15, 10),color='pink')
plt.tight_layout()
plt.show()
```

```python
smoker_counts = df['smoker'].value_counts()
plt.figure(figsize=(10, 6))
plt.barh(smoker_counts.index, smoker_counts.values, color='skyblue')
plt.xlabel('Count')
plt.ylabel('Smoker')
plt.show()
```

```python
sex_counts = df['sex'].value_counts()

# Pasta grafii izme
plt.figure(figsize=(8, 8))
plt.pie(sex_counts.values, labels=sex_counts.index, autopct='%1.1f%%', startangle=140,explode=[0,0.1], shadow=True)
plt.title('Distribution of Male and Female')
plt.show()
```

```python
region_counts = df['region'].value_counts()

# Pasta grafii izme
plt.figure(figsize=(8, 8))
plt.pie(region_counts.values, labels=region_counts.index, autopct='%1.1f%%', startangle=140,explode=[0,0,0,0.1], shadow=True)
plt.title('Distribution of Male and Female')
plt.show()
```

```python
#Cinsiyet Dalm
plt.figure(figsize=(10,6))
sns.countplot(data=df,x="sex",color='red',hue="sex",palette={'female': 'pink', 'male': 'lightblue'})
plt.title("Cinsiyet Dalm")
plt.show()
```
